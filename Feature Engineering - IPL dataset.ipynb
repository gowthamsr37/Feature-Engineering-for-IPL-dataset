```python
df['isDot'] = df['runs_off_bat'].apply(lambda x: 1 if x == 0 else 0)
df['isOne'] = df['runs_off_bat'].apply(lambda x: 1 if x == 1 else 0)
df['isTwo'] = df['runs_off_bat'].apply(lambda x: 1 if x == 2 else 0)
df['isThree'] = df['runs_off_bat'].apply(lambda x: 1 if x == 3 else 0)
df['isFour'] = df['runs_off_bat'].apply(lambda x: 1 if x == 4 else 0)
df['isSix'] = df['runs_off_bat'].apply(lambda x: 1 if x == 6 else 0)
```

```python
df['over_num'] = df['ball'].astype('str').str.split('.').str.get(0).astype('int')+1
df['ball_num'] = df['ball'].astype('str').str.split('.').str.get(1).astype('int')
```

```python
def phase_of_play(over_num):
    if over_num <=6:
        return 'Powerplay'
    elif over_num <=15:
        return 'Middle'
    else:
        return 'Death'
```

```python
df['Phase_of_play'] = df['over_num'].apply(lambda x: phase_of_play(x))
```

```python
def is_bowlers_wicket(player_dismissed, wicket_type):
    if type(player_dismissed) == str:
        if wicket_type not in [ 'run out', 'retired hurt', 'obstructing the field']:
            return 1
        else:
            return 0
    else:
        return 0
```

```python
df['isBowlerWk'] = df.apply(lambda x: is_bowlers_wicket(x['player_dismissed'], x['wicket_type']), axis = 1)
```

```python
df.head()
```

```python
df['runs'] = df['runs_off_bat'] + df['extras']
```

```python
df['current_score'] = df.groupby(['match_id' , 'innings'])['runs'].cumsum()
df['Total_score'] = df.groupby(['match_id' , 'innings'])['runs'].transform('sum')
```

```python
df['runs_not_counting_for_maiden_overs'] = df['runs_off_bat'] + df['wides'] + df['noballs']
```

```python
df['wides'] = df['wides'].fillna(0)
df['noballs'] = df['noballs'].fillna(0)
```

```python
df['runs_not_counting_for_maiden_overs'] = df['runs_off_bat'] + df['wides'] + df['noballs']
```
